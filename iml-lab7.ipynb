```python
#Bagthariya Mahek R.
#216270307080
#Practical No.7: Write a Pandas program to implement following operation
#to write a DataFrame to CSV file using tab separator

import numpy as np
import pandas as pd
students = {'Student': ['Mahek', 'Dhruvi','Jal', 'Dhruti'],'RollNumber': [80, 90, 82, 16],'Grade': ['A', 'C', 'F', 'B']}
df = pd.DataFrame(students,columns =['Student', 'RollNumber','Grade'])

print("Original DataFrame")
print(df)

df.to_csv('Students.csv', sep ='\t')
new_df = pd.read_csv('Students.csv')


print('Data from Students.csv:')
print(new_df)
```
