```python
#Bagthariya Mahek r.
#216270307080
#Practical No.5: Write a NumPy program to implement following operation
#to find the maximum and minimum value of a given flattened array

import numpy as np
flattened_array = [1, 6,8,22,56,77,43,44,90]
# Convert the list to a NumPy array
np_array = np.array(flattened_array)

# Use NumPy's built-in functions to find max and min
max_val = np.max(np_array)
min_val = np.min(np_array)

print(f"Maximum value: {max_val}")
print(f"Minimum value: {min_val}")
```
