```python
#Bagthariya Mahek r.
#216270307080
#Practical No.12: Write a Python program using Scikit-learn to print the keys, number of rows-columns, feature names and the description of the given data

from sklearn import datasets

iris = datasets.load_breast_cancer()
#df = datasets.read_csv('/kaggle/input/nifty50-stock-market-data/KOTAKBANK.csv')

print("Keys :")
print(iris.keys())

n_rows, n_cols = iris.data.shape
print(f"\n\n\n\nNumber of Rows :{n_rows}, Number of columns: {n_cols}")

print("\n\n\n\nFeature Names :")
print(iris.feature_names)

print("\n\n\n\nDescription :")
print(iris.DESCR)
```
