```python
#Bagthariya Mahek r.
#216270307080
#Practical No.3: Write a NumPy program to implement following operation
#to stack arrays horizontally (column wise)

import numpy as np
x = np.array([[6, 7, 8],
            [8, 8, 8]])
y = np.array([[9, 8, 7],
                 [2, 3, 1]])
print("\nx: \n", x)
print("\ny: \n", y)

# horizontally stack the arrays (axis = 1)

z = np.hstack([x, y])
print("\nz: \n", z)
```
